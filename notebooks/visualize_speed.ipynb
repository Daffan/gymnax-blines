```python

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%autoreload 2

import matplotlib as mpl
mpl.rcParams['figure.dpi']= 300

import sys
sys.path.append("..")
from utils.helpers import load_pkl_object

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("poster")
```

```python
env_names = ['Acrobot-v1', 'Pendulum-v1', 'CartPole-v1', 'MountainCar-v0', 'MountainCarContinuous-v0',
             #'Asterix-MinAtar', 'Breakout-MinAtar', 'Freeway-MinAtar', 'Seaquest-MinAtar', 'SpaceInvaders-MinAtar'
             ]
labels = ["np-10-rand-cpu", "np-10-mlp-cpu",
          "jax-10-rand-cpu", "jax-10-mlp-cpu",
          "np-40-rand-cpu", "np-40-mlp-cpu",
          "jax-40-rand-cpu", "jax-40-mlp-cpu",
          "jax-2k-rand-2080Ti", "jax-2k-mlp-2080Ti",
#          "jax-2k-rand-2080Ti", "jax-2k-mlp-2080Ti",
#          "jax-2k-rand-2080Ti", "jax-2k-mlp-2080Ti",
          ]
c = ['blue', 'blue', 'yellow', 'yellow', 'blue', 'blue', 'yellow', 'yellow', 'yellow', 'yellow']

fig, axs = plt.subplots(figsize=(20, 20), nrows=3, ncols=3)


for i, e in enumerate(env_names):
    results = load_pkl_object(f"../agents/{e}/speed.pkl")
    mean_results, std_results = [], []
    for k in results.keys():
        mean_results.append(results[k][0])
        std_results.append(results[k][1])
    ax = axs.flatten()[i]
    
    hbars = ax.barh(np.arange(len(labels)), mean_results, xerr=std_results, color=c)
    
    if i in [0, 3, 6]:
        ax.set_yticks(np.arange(len(labels)), labels=labels)
    else:
        ax.set_yticks([], [])
    #ax.set_xlim(0, 10)
    ax.invert_yaxis()  # labels read top-to-bottom
    if i in [6, 7, 8]:
        ax.set_xlabel('Secs/1M Step Transitions')
    ax.set_title(f'{e}')

    # Label with specially formatted floats
    ax.bar_label(hbars, fmt='%.2f', label_type='edge')

fig.tight_layout()
```

```python
for i, e in enumerate(env_names):
    results = load_pkl_object(f"../agents/{e}/speed.pkl")
    l0 = list(results.keys())[0]
    l1 = list(results.keys())[2]
    l2 = list(results.keys())[1]
    l3 = list(results.keys())[3]
    speed_up_rand = results[l0][0] / results[l1][0]
    speed_up_mlp = results[l2][0] / results[l3][0]
    print(r"{} - Random {:.2f} - MLP {:.2f}".format(e, speed_up_rand, speed_up_mlp))
```
